```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.xkcd()  # Yes...
plt.rcdefaults()
```

```python
king = [0,0,8.9,7.9]
queen = [0,0,9.9,6.9]
man = [0,0,0.97,6.9]
woman = [0,0,1.97,5.9]

X, Y, U, V = zip(*np.array([king, queen, man, woman]))

fig, ax = plt.subplots(figsize=(10,6))
ax = plt.gca()

ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, width=0.002)


# ax.plot(9,8,'bo')

ax.plot(9,8,'ro')
ax.text(8.7, 8.3,'king')
ax.plot(10,7,'ro')
ax.text(9.7, 7.3,'queen = king - man + woman')
ax.plot(1,7,'ro')
ax.text(0.7, 7.3,'man')
ax.plot(2,6,'ro')
ax.text(1.7, 6.3,'woman')


ax.plot([9,8], [8,1], linestyle="--", color="red")
ax.plot([8,10], [1,7], "--", color="red")
ax.plot(8,1, "ro")
ax.text(8.2,1, "king - man")
ax.text(9.2, 4.2, "+ woman")

ax.set_xlim([0, 10.2])
ax.set_ylim([0, 10])
# plt.draw()

plt.xticks(np.arange(11))
plt.yticks(np.arange(11))
plt.title('Vectors de paraules - Exemple "queen = king - man + woman"')
# ax.xaxis.grid(True, which='minor')
ax.tick_params(which='major', width=1)
# Show plot with grid
ax.grid(linewidth=0.2)
# ax.minorticks_on()

sns.despine()



plt.show()

# plt.show()
```
