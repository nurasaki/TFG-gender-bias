```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()

data = pd.DataFrame({
    "word":np.arange(1,11),
    "genere": (["femení"]*2 + ["masculí"]*8),
    "P_uniform": np.repeat(1/10, 10),
    "P_fem": [1/2]*2+[0]*8,
    "P_masc": [0]*2+[1/8]*8
})

data['odds'] = (data["P_fem"] + data["P_masc"])/data["P_uniform"]
data['log_odds'] = np.log(data['odds'])

fig, axs = plt.subplots(1,4, sharey=True, figsize=(12,4))
# fig, axs = plt.subplots(1,4, figsize=(12,4))
plt.subplots_adjust(hspace=0.1, wspace=0.13)


cols = ["P_uniform", "P_fem", "P_masc", "log_odds"]
# cols = ["P_uniform", "P_fem", "P_masc", "odds"]
titols = ["Distribució prob. uniforme","Distribució prob. (context femení)", 
          "Distribució prob. (context masculí)","Associació (Log Odds ratio)"]
fontsize="small"

for ax, titol, col in zip(axs, titols, cols):

    for g in ["femení", "masculí"]:
        
        values = data.apply(lambda r: r[col] if r.genere==g else 0, axis=1)
        sns.barplot(y=values, x=data.word, color=sns.color_palette()[g=="masculí"], ax=ax, label=g)

    ax.set_ylabel("")
    ax.set_xlabel("")
    # ax.set_xticklabels(fontsize=6)
    ax.set_title(titol, fontsize=fontsize)

    # ax.xticks(fontsize=10)
    ax.xaxis.set_tick_params(labelsize=8)
    
axs[0].set_ylabel("Probabilitat")
axs[3].set_ylabel("Log Odds")

axs[0].legend(fontsize="xx-small",loc='upper left')

plt.show()
```
